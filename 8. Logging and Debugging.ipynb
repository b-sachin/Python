### Exercise

1. Create a logger in your code 
2. create one function which can take any number of input as an argument and it will be able to return sum of it.
3. capture user input in log
4. give user instruction in log file
5. read a log file from python code and show all the log in console.

```python
# 1. Create a logger in your code 

import logging as lg
import os

os.mkdir("Logger")
os.chdir(os.getcwd()+"\\Logger")
lg.basicConfig(filename = "logger.log", level=lg.INFO)


# 2. create one function which can take any number of input as an argument and it will be able to return sum of it.

def add(*args):
    # 4. give user instruction in log file
    lg.info("Inside add(*args) function")
    res = 0
    for i in args:
        # 3. capture user input in log
        lg.info(i)
        res += i
    lg.info("Result generated: "+str(res))
    return res

add(10,20,30,40,50)

# 5. read a log file from python code and show all the log in console.
f = open(os.getcwd()+"\\logger.log","r")
print(f.read())
```

### Logging Complete Code 

```python
import logging as lg

lg.basicConfig(filename = 'app.log', level=lg.DEBUG, format= '%(name)s %(asctime)s %(levelname)s %(message)s')

#create handlers
console_log = lg.StreamHandler()
console_log.setLevel(lg.DEBUG)
format = lg.Formatter('%(name)s %(asctime)s %(levelname)s %(message)s')
console_log.setFormatter(format)

#create a custom logger

lg.getLogger('').addHandler(console_log)

lg.info('main log')

lg1 = lg.getLogger('user1')
lg2 = lg.getLogger('user2')

lg1.info('this log is from user1 info')
lg2.debug('this log is from user2 debug')
```
